```python
# read master data 
master_df = pd.read_csv('master_data_p1_p3.csv',index_col=0)
master_df.head()
```

# PART 4 ( preprocess all remaining csv files)

```python
# read previous application csv file 
previous = pd.read_csv('data/previous_application.csv')
previous = convert_types(previous, print_info=True)
previous.head()
```

```python
# Calculate aggregate statistics for each numeric column
previous_agg = agg_numeric(previous, 'SK_ID_CURR', 'previous')
print('Previous aggregation shape: ', previous_agg.shape)
previous_agg.head()
```

```python
# Calculate value counts for each categorical column
previous_counts = agg_categorical(previous, 'SK_ID_CURR', 'previous')
print('Previous counts shape: ', previous_counts.shape)
previous_counts.head()
```

```python
# Merge to Master dataframe
master_df = master_df.merge(previous_counts, on ='SK_ID_CURR', how = 'left')
master_df = master_df.merge(previous_agg, on = 'SK_ID_CURR', how = 'left')
master_df.head()
master_df.shape

gc.enable()
del previous_counts, previous_agg, previous
gc.collect()
```

```python
master_df_missing = remove_missing_columns(master_df)
```

### Cash data

```python
# read cash data
cash = pd.read_csv('data/POS_CASH_balance.csv')
cash = convert_types(cash, print_info=True)
cash.head()
```

```python
cash_by_client = aggregate_client(cash, group_vars = ['SK_ID_PREV', 'SK_ID_CURR'], df_names = ['cash', 'client'])
cash_by_client.head()
```

```python
print('Cash by Client Shape: ', cash_by_client.shape)
master_df = master_df.merge(cash_by_client, on = 'SK_ID_CURR', how = 'left')

gc.enable()
del cash, cash_by_client
gc.collect()

master_df.head()
master_df.shape
```

```python
credit = pd.read_csv('data/credit_card_balance.csv')
credit = convert_types(credit, print_info = True)
credit.head()
```

```python
credit_by_client = aggregate_client(credit, group_vars = ['SK_ID_PREV', 'SK_ID_CURR'], df_names = ['credit', 'client'])
credit_by_client.head()
```

```python
print('Credit by client shape: ', credit_by_client.shape)

master_df = master_df.merge(credit_by_client, on = 'SK_ID_CURR', how = 'left')

gc.enable()
del credit, credit_by_client
gc.collect()

master_df.head()
master_df.shape
```
